```python
worker_class_df = pd.DataFrame()
```

```python
worker_class_list = ['Employee of private company workers', 'Self-employed in own incorporated business workers','Private not-for-profit wage and salary workers', 
                    'Local government workers', 'State government workers', 'Federal government workers', 'Self-employed in own not incorporated business workers and unpaid family workers']
```

```python
for class_worker_index in range(len(worker_class_list)):  # Loop over the index of industry_list
    class_worker_int = str(class_worker_index + 3).zfill(2)  # Construct industry_int based on industry_index
    try:
        class_data = pd.DataFrame()  # Initialize as an empty DataFrame for each industry
        print(f"Fetching data for class of workers {worker_class_list[class_worker_index]}...")
        for year in range(2005, 2023):
            print(f"Fetching data for year {year}...")
            url = f'https://api.census.gov/data/{year}/acs/acs1?get=NAME,B24081_0{class_worker_int}E&for=place:*&key={census_api}'
            response = requests.get(url)
            if response.status_code == 200:
                json_response = response.json()
                data = pd.DataFrame(json_response[1:], columns=json_response[0])
                data = data.rename(columns={f'B24081_0{class_worker_int}E': 'Income'})
                data = data.dropna()
                data = data.loc[data['Income'] != '0']
                data['Class of Worker'] = worker_class_list[class_worker_index]
                data['Income'] = data['Income'].astype(int)
                data = data.sort_values('Income', ascending=False)
                data['Year'] = str(year)
                class_data = pd.concat([class_data, data], ignore_index=True)  # Concatenate data for each year
                print(f"Data for year {year} fetched successfully.")
                print(f"Size of class of workers data: {class_data.shape}")
                print(f"Size of data after concatenation: {data.shape}")
            else:
                print(f"Error fetching data for year {year}: Status code {response.status_code}")
        worker_class_df = pd.concat([worker_class_df, class_data], ignore_index=True)  # Concatenate data for each industry
        print(f"Size of combined data: {worker_class_df.shape}")
    except Exception as e:
        print(f'Error fetching data for females in industry {worker_class_list[class_worker_index]}: {e}')
print(worker_class_df)
```

```python
government_income = worker_class_df.loc[(worker_class_df['Class of Worker'] == 'Local government workers') | (worker_class_df['Class of Worker'] == 'State government workers') | (worker_class_df['Class of Worker'] == 'Federal government workers')]
```

```python
government_income['Year'] = government_income['Year'].astype(str)
```

```python
worker_class_dataframe = government_income.groupby(['NAME', 'Year', 'Class of Worker']).sum().reset_index()
```

```python
worker_class_dataframe
```

```python
for index, row in worker_class_dataframe.iterrows():
    if row['Class of Worker'] == 'State government workers':
        cumulative_income = (row['Income'] + worker_class_dataframe.loc[index - 1, 'Income'] + worker_class_dataframe.loc[index - 2, 'Income']) / 3
        worker_class_dataframe.at[index, 'Income'] = cumulative_income
```

```python
worker_class_dataframe = worker_class_dataframe.loc[worker_class_dataframe['Class of Worker'] == 'State government workers']
```

```python
worker_class_dataframe['Class of Worker'] = 'Local, state, and federal government workers'
```

```python
worker_class_dataframe = worker_class_dataframe[['NAME', 'Income', 'state', 'place', 'Class of Worker', 'Year']]
```

```python
type_of_worker = worker_class_df.loc[~((worker_class_df['Class of Worker'] == 'Local government workers') | (worker_class_df['Class of Worker'] == 'State government workers') | (worker_class_df['Class of Worker'] == 'Federal government workers'))]
```
