```python
import pandas as pd
from pathlib import Path
from functools import reduce
```

```python
zips_2014 = Path('us_zip_pop_2014.csv')
zips_2015 = Path('us_zip_pop_2015.csv')
zips_2016 = Path('us_zip_pop_2016.csv')
zips_2017 = Path('us_zip_pop_2017.csv')
zips_2018 = Path('us_zip_pop_2018.csv')
zips_2019 = Path('us_zip_pop_2019.csv')
zips_2020 = Path('us_zip_pop_2020.csv')
zips_2021 = Path('us_zip_pop_2021.csv')
zips_2022 = Path('us_zip_pop_2022.csv')
zip_city = Path('uszips.csv')
```

```python
columns_to_use = ['NAME', 'B01003_001E']
columns_in_zip = ['zip', 'lat', 'lng', 'city', 'state_id', 'state_name', 'county_name', 'timezone']
```

```python
pop_2014 = pd.read_csv(zips_2014, usecols = columns_to_use)
pop_2015 = pd.read_csv(zips_2015, usecols = columns_to_use)
pop_2016 = pd.read_csv(zips_2016, usecols = columns_to_use)
pop_2017 = pd.read_csv(zips_2017, usecols = columns_to_use)
pop_2018 = pd.read_csv(zips_2018, usecols = columns_to_use)
pop_2019 = pd.read_csv(zips_2019, usecols = columns_to_use)
pop_2020 = pd.read_csv(zips_2020, usecols = columns_to_use)
pop_2021 = pd.read_csv(zips_2021, usecols = columns_to_use)
pop_2022 = pd.read_csv(zips_2022, usecols = columns_to_use)
city_zip = pd.read_csv(zip_city, usecols = columns_in_zip)
```

```python
pop_2014 = pop_2014.iloc[1:]
pop_2014 = pop_2014.rename(columns={'NAME': 'zip', 'B01003_001E': '2014'})
pop_2015 = pop_2015.iloc[1:]
pop_2015 = pop_2015.rename(columns={'NAME': 'zip', 'B01003_001E': '2015'})
pop_2016 = pop_2016.iloc[1:]
pop_2016 = pop_2016.rename(columns={'NAME': 'zip', 'B01003_001E': '2016'})
pop_2017 = pop_2017.iloc[1:]
pop_2017 = pop_2017.rename(columns={'NAME': 'zip', 'B01003_001E': '2017'})
pop_2018 = pop_2018.iloc[1:]
pop_2018 = pop_2018.rename(columns={'NAME': 'zip', 'B01003_001E': '2018'})
pop_2019 = pop_2019.iloc[1:]
pop_2019 = pop_2019.rename(columns={'NAME': 'zip', 'B01003_001E': '2019'})
pop_2020 = pop_2020.iloc[1:]
pop_2020 = pop_2020.rename(columns={'NAME': 'zip', 'B01003_001E': '2020'})
pop_2021 = pop_2021.iloc[1:]
pop_2021 = pop_2021.rename(columns={'NAME': 'zip', 'B01003_001E': '2021'})
pop_2022 = pop_2022.iloc[1:]
pop_2022 = pop_2022.rename(columns={'NAME': 'zip', 'B01003_001E': '2022'})
```

```python
city_zip['zip'] = city_zip['zip'].astype(str).str.zfill(5)
city_zip
```

```python
pop_2014['zip'] = pop_2014['zip'].str.replace("ZCTA5 ", '')
pop_2015['zip'] = pop_2015['zip'].str.replace("ZCTA5 ", '')
pop_2016['zip'] = pop_2016['zip'].str.replace("ZCTA5 ", '')
pop_2017['zip'] = pop_2017['zip'].str.replace("ZCTA5 ", '')
pop_2018['zip'] = pop_2018['zip'].str.replace("ZCTA5 ", '')
pop_2019['zip'] = pop_2019['zip'].str.replace("ZCTA5 ", '')
pop_2020['zip'] = pop_2020['zip'].str.replace("ZCTA5 ", '')
pop_2021['zip'] = pop_2021['zip'].str.replace("ZCTA5 ", '')
pop_2022['zip'] = pop_2022['zip'].str.replace("ZCTA5 ", '')
```

```python
pop_2020
```

```python
pop_dataframes = [pop_2022, pop_2021, pop_2020, pop_2019, pop_2018, pop_2017, pop_2016, pop_2015, pop_2014]
population_zip = reduce(lambda left, right: pd.merge(left, right, on='zip'), pop_dataframes)
column_order = ['2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']

population_zip.set_index('zip', inplace=True)
population_zip = population_zip[column_order]
population_zip
```

```python
city_and_zip = pd.merge(population_zip, city_zip, on='zip')
city_and_zip.set_index('zip', inplace=True)
city_and_zip
```
